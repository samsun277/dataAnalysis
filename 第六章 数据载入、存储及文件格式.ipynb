#### 如何处理大型文本文件
- 调整显示参数：pd.options.display.min_rows = 50
- 参数nrows:限制从文件头读入的行数
- chunksize：

```python
# pd.options.display.max_rows = 300
# DataFrame会按照min_rows指定的行数进行显示
pd.options.display.min_rows = 10
df = pd.read_csv('Sample/MC30_AllFB_20210515.csv', header = None, names = columns)
```

```python
chunker = pd.read_csv('Sample/MC30_AllFB_20210515.csv', header = None, names = columns, chunksize = 1000)
price_max = []
for piece in chunker:
   price_max.append(piece['Price'].max())
price_max
```
