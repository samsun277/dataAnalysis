```python
trades = meituan[meituan['Message Type']==50]
#df['match'] = df.col1.eq(df.col1.shift())
trades['B/S'] = trades['Price'].gt(trades['Price'].shift())
trades[trades['Price'].eq(trades['Price'].shift())] = np.nan
trades[trades['B/S'].notnull()]
```

```python
#meituan_rise_1 = meituan[(meituan['Date/Time'] > 20220301141000000) & (meituan['Date/Time'] < 20220301151000000) & (meituan['Message Type'].isin([50,30,31,32]))]
meituan_rise_1 = meituan[(meituan['Date/Time'] > 20220301093000000) & (meituan['Date/Time'] < 20220301140000000) & (meituan['Message Type'].isin([50,30,31,32]))]
meituan_rise_1
```

```python
from datetime import datetime

def convert_to_datetime(a):
    return datetime.strptime(str(a), '%Y%m%d%H%M%S%f')

datetime_obj = meituan_rise_1['Date/Time'].apply(convert_to_datetime)

meituan_rise_1.insert(loc=0, column='RealDateTime', value = datetime_obj)
meituan_rise_1.groupby(pd.Grouper(key = 'RealDateTime', freq ='10s')).count()
```

```python
order_counts = meituan_rise_1.groupby([pd.Grouper(key = 'RealDateTime', freq ='10s'), 'Message Type'])['Date/Time'].count()
```

```python
order_counts_frame = order_counts.unstack()
```

```python
import matplotlib.dates as md
fig, ax = plt.subplots(1,1)
#axes.legend(['加单','改单','删单','交易'])
ax.xaxis.set_major_locator(md.MinuteLocator(byminute = np.arange(60)))
ax.xaxis.set_major_formatter(md.DateFormatter('%H:%M'))
plt.setp(ax.xaxis.get_majorticklabels(), rotation = 90)
order_counts_frame.plot(ax = ax, figsize=(200,50), x_compat = True)
fig.savefig("foo.png")
```
